```python
library(affy)
library(dplyr)
```

```python
# 1. Load CEL files
data <- ReadAffy(celfile.path = "../data/CEL_files")
```

```python
# 2. Normalize using RMA
eset <- rma(data)
```

```python
# 3. Extract expression matrix
exprs_matrix <- exprs(eset)
```

```python
# 4. Convert to data frame
exprs_df <- as.data.frame(exprs_matrix)
```

```python
# 5. Add gene IDs as a column
exprs_df$probe_id <- rownames(exprs_df)
```

```python
# 6. Move probe_id to first column
exprs_df <- exprs_df %>%
  relocate(probe_id)
```

```python
# 7. Save to CSV
write.csv(
    exprs_df,
    file = "../data/GSE42568.csv",
    row.names = FALSE
)
```
